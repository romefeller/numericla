```python
def f(x):
  return 2*x-7

def g(x):
  return 0
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

points = [-7.1, -4.5, -3.1, -0.5, 0.9, 3.5, 4.9, 7.5, 8.9, 11.5]
color = ["b","g","g","g","g","g","r","r","r","r"]

df_aux = pd.DataFrame({"x" : range(10), "y" : points, "color" : color})
df = pd.DataFrame({"x" : range(10), "y" : points}) # 1
covM = df.cov() # 2
print(covM)

eigenValues, eigenVectors = np.linalg.eig(covM) # 3

sorted_indices = np.argsort(-eigenValues) # 4

sorted_eigenValues = eigenValues[sorted_indices] # 4

sorted_eigenVectors = eigenVectors[:, sorted_indices] # 4

print("Autovalores: ", sorted_eigenValues)
print("Autovectores: \n", sorted_eigenVectors)

principal_eigenVector = sorted_eigenVectors[:, 0] # 4

projection = df.dot(principal_eigenVector)

projection_df = pd.DataFrame(projection, columns=['PCA'])

x = np.linspace(-3,10,1000)
plt.figure(figsize=(8, 6))
plt.plot()
plt.scatter(df['x'], df['y'],c=df_aux['color'])
plt.plot(x,f(x))
plt.show()

x = np.linspace(-15,10,1000)
plt.figure(figsize=(8, 6))
plt.scatter(projection, np.zeros_like(projection),c=df_aux['color'])
plt.plot(x,g(x))
plt.title('Principal Eigenvector')
plt.xlabel('Projection Value')
plt.legend()
plt.show()

```
